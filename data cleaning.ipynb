```python
import json
from os import path
import pandas as pd
user = []
business=[]
review=[]
for line in open('/Users/hyy/Desktop/Columbia/E 4571/Project/yelp_training_set/yelp_training_set_user.json', 'r'):
    user.append(json.loads(line))
for line in open('/Users/hyy/Desktop/Columbia/E 4571/Project/yelp_training_set/yelp_training_set_business.json', 'r'):
    business.append(json.loads(line))
for line in open('/Users/hyy/Desktop/Columbia/E 4571/Project/yelp_training_set/yelp_training_set_review.json', 'r'):
    review.append(json.loads(line))
```

```python
restaurants=[]
for i in business:
    if 'Restaurants'in i['categories'] and i['open']==True and i['city']=='Phoenix'and i['review_count']>95:
        restaurants.append(i)
```

```python
id=[]
for i in restaurants:
    if i['business_id'] not in id:
        id.append(i['business_id'])
```

```python
user_id=[]
for i in review:
    if i['business_id'] in id:
        if i['user_id'] not in user_id:
            user_id.append(i['user_id'])
```

```python
df= pd.DataFrame(index=user_id, columns=id)
```

```python
for i in review:
    try:
        if i['user_id'] in user_id:
            x=i['business_id']
            y=i['user_id']
            df[x][y]=i['stars']
    except:
        pass
```

```python
df=df.dropna(thresh=2)
```

```python
user_id=list(df.index)
ave_rate={}
for i in user:
    if i['user_id'] in user_id:
        ave_rate[i['user_id']]=i['average_stars']
```

```python
new_list=list(ave_rate.keys())
df=df.ix[new_list]
ave_df= pd.DataFrame([ave_rate],columns=ave_rate.keys()).transpose()
ave_df.columns=['average_rate']
```

```python
df.to_csv('rating data 95.csv', encoding='utf-8',index=True)
ave_df.to_csv('average rating data 95.csv', encoding='utf-8',index=True)
```
